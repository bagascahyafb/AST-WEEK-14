# Menampilkan Gambar

```python
plt.figure(figsize=(24, 12))
for i in range(1, 11):
    nama_file = f'minato{i}.jpeg'
    citra = cv2.imread(nama_file)

    if citra is not None:
        plt.subplot(2, 5, i)  # Ubah angka 2 dan 5 sesuai dengan jumlah baris dan kolom subplot yang diinginkan
        plt.imshow(cv2.cvtColor(citra, cv2.COLOR_BGR2RGB))
        plt.title(f'Minato ke-{i}')
        plt.axis('off')  # Optional: Nonaktifkan sumbu x dan y
    else:
        print(f'File {nama_file} tidak ditemukan.')

plt.show()
```

<b>Analisis:</b>

<div>
    <p style="text-align: justify;">Untuk menampilkan gambar yang telah dimiliki, adalah dengan menggunakan fungsi imread yang berisi parameter path dari gambar yang dimiliki, pada kode diatas dilakukan perulangan agar semua gambar yang dimiliki dapat ditampilkan, akan tetapi, sebelum ditampilkan, dilakukan conditional statement dimana apabila path yang dimaksud tidak ada, maka akan tercetak "File {nama_file} tidak ditemukan.
    </p>
<div>



```python
for i in range(1, 11):
    nama_file = f'minato{i}.jpeg'
    citra = cv2.imread(nama_file)

    if citra is not None:
        cv2.imwrite(f'minato{i}resize.jpeg', citra)
    else:
        print(f'File {nama_file} tidak ditemukan.')
```

# Grayscale Images

```python
plt.figure(figsize=(24, 12))

for i in range(1, 11):
    nama_file = f'minato{i}.jpeg'
    img = cv2.imread(nama_file)

    if img is not None:
        plt.subplot(4, 10, i)  # Adjust the number of rows and columns as needed
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f'Minato ke-{i}')
        plt.axis('off')  # Optional: Turn off x and y axes

        # Grayscaling using different methods
        grey_1 = np.clip(0.299 * img[:, :, 0] + 0.587 * img[:, :, 1] + 0.114 * img[:, :, 2], 0, 255).astype(np.uint8)
        grey_2 = np.clip(0.333 * img[:, :, 0] + 0.333 * img[:, :, 1] + 0.333 * img[:, :, 2], 0, 255).astype(np.uint8)
        grey_3 = np.clip(0.07 * img[:, :, 0] + 0.72 * img[:, :, 1] + 0.21 * img[:, :, 2], 0, 255).astype(np.uint8)

        plt.subplot(4, 10, i + 10)
        plt.imshow(grey_1, cmap='gray')
        plt.title(f'Grayscale-1')
        plt.axis('off')

        plt.subplot(4, 10, i + 20)
        plt.imshow(grey_2, cmap='gray')
        plt.title(f'Grayscale-2')
        plt.axis('off')

        plt.subplot(4, 10, i + 30)
        plt.imshow(grey_3, cmap='gray')
        plt.title(f'Grayscale-3')
        plt.axis('off')

    else:
        print(f'File {nama_file} tidak ditemukan.')

# Show the entire plot
plt.show()
```

<b>Analisis:</b>

<div>
    <p style="text-align: justify;">Dilakukan perubahan warna pada data yang ada, merubah warnanya menjadi Gray Scale menggunakan 3 rumus, diantaranya adalah:<br> 
    grayscale1 = 0.299R + 0.587G + 0.114B<br>
    grayscale2 = 0.333R + 0.333G + 0.333B<br>
    grayscale3 = 0.07R + 0.72G + 0.21B<br>
    Berdasarkan hasilnya, dapat dilihat bahwa tidak ada perbedaan yang mencolok dari Gray Scale 1 dengan 2, akan tetapi pada grayscale 3, terlihat warna yang dihasilkan lebih terang, yang terlihat jelas pada warna rambut dari karakter tersebut.
